```python
import os
import pycolmap
from tqdm import tqdm
from pathlib import Path
from suppressor.DSEC_dataloader.provider import DatasetProvider


dsec_dir = "/home/rpg/Downloads/DSEC"

dataset_provider = DatasetProvider(dsec_dir, num_bins=1, representation='voxel')
dataset = dataset_provider.get_test_dataset()
```

```python
for sequence in tqdm(dataset):
    image_dir = sequence.images_path
    sequence_path = Path(sequence.sequence_path)

    output_path = Path(sequence_path) / "colmap"
    database_path = output_path / "database.db"

    # Create output directories
    output_path.mkdir(parents=True, exist_ok=True)

    # Create the COLMAP database
    database = pycolmap.Database(database_path)
    database.close()


    pycolmap.extract_features(database_path, image_dir, sift_options={"max_num_features": 150})
    pycolmap.match_exhaustive(database_path)
    maps = pycolmap.incremental_mapping(database_path, image_dir, output_path)
    maps[0].write(output_path)

    break
```
